```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-20, 21, 1, dtype=np.float32)
y = x + 10

x_train, y_train = x[:35], y[:35]
x_test, y_test = x[35:], y[35:]

model = tf.keras.Sequential ([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(1)
])
model.compile(
    loss = "mae",
    optimizer = tf.keras.optimizers.SGD(learning_rate=0.03),
    metrics = ["mae"]
)

history = model.fit(x_train, y_train, epochs=200, verbose=1)
loss, mae = model.evaluate(x_test, y_test, verbose=0)
print("Test MAE:", mae)

y_pred = model.predict(x_test).squeeze()
plt.figure(figsize=(9,6))
plt.scatter(x_train, y_train, label="Train data", c="blue")
plt.scatter(x_test, y_test, label="Test data", c="green")
plt.scatter(x_test, y_pred, label="Predictions", c="red", marker = "X")
plt.legend(); plt.title("Regression Predictions")
plt.show()
```
